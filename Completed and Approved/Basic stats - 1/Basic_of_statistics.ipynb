```python
# prompt: Load the dataset = "/content/sales_data_with_discounts.csv" into a data analysis tool or programming environment (e.g., Python with pandas library).

import pandas as pd

# Load the dataset into a pandas DataFrame
sales_data  = pd.read_csv(r"D:\EXCELR\Assignments\Basic stats - 1\sales_data_with_discounts.csv")

# Print the first few rows of the DataFrame
print(sales_data.head())
```

```python
# prompt: Identify numerical columns in the dataset.

numerical_columns = sales_data.select_dtypes(include=["int64", "float64"]).columns
print(f"Numerical columns: {numerical_columns.tolist()}")
```

```python
# prompt: Calculate the mean, median, mode, and standard deviation for these columns.
for column in numerical_columns:
  mean = sales_data[column].mean()
  median = sales_data[column].median()
  mode = sales_data[column].mode()[0]
  std_dev = sales_data[column].std()
  print(f"Column: {column}")
  print(f"Mean: {mean}")
  print(f"Median: {median}")
  print(f"Mode: {mode}")
  print(f"Standard deviation: {std_dev}")
  print("")
```

Data Visualization

```python
# prompt: To visualize the distribution and relationship of numerical and categorical variables in the dataset. Please give summary for each plot as well for each

import matplotlib.pyplot as plt

# Histogram for each numerical column
for column in numerical_columns:
  plt.hist(sales_data[column], bins=10)
  plt.title(f"Histogram of {column}")
  plt.xlabel(column)
  plt.ylabel("Frequency")
  plt.show()

# Boxplot for each numerical column
for column in numerical_columns:
  plt.boxplot(sales_data[column])
  plt.title(f"Boxplot of {column}")
  plt.xlabel(column)
  plt.ylabel("Value")
  plt.show()

# Scatter plot for each pair of numerical columns
for i, column1 in enumerate(numerical_columns):
  for j, column2 in enumerate(numerical_columns):
    if i < j:
      plt.scatter(sales_data[column1], sales_data[column2])
      plt.title(f"Scatter plot of {column1} vs. {column2}")
      plt.xlabel(column1)
      plt.ylabel(column2)
      plt.show()

# Bar chart for each categorical column
categorical_columns = sales_data.select_dtypes(include=["object"]).columns
for column in categorical_columns:
  value_counts = sales_data[column].value_counts()
  plt.bar(value_counts.index, value_counts.values)
  plt.title(f"Bar chart of {column}")
  plt.xlabel(column)
  plt.ylabel("Count")
  plt.show()

# Pie chart for each categorical column
for column in categorical_columns:
  value_counts = sales_data[column].value_counts()
  plt.pie(value_counts.values, labels=value_counts.index)
  plt.title(f"Pie chart of {column}")
  plt.show()

# Summary for each plot:

# Histograms: Show the distribution of values in each numerical column.
# Boxplots: Show the distribution of values in each numerical column, including the median, quartiles, and outliers.
# Scatter plots: Show the relationship between two numerical columns.
# Bar charts: Show the frequency of each category in a categorical column.
# Pie charts: Show the proportion of each category in a categorical column.
```

```python
# prompt: Explain the concept of standardization (z-score normalization) for above numerical columns

# Standardization (z-score normalization) is a technique for scaling numerical columns to have a mean of 0 and a standard deviation of 1. This makes it easier to compare values from different columns and to identify outliers.

# To standardize a column, we first subtract the mean from each value and then divide by the standard deviation.

# In the code above, we first calculate the mean and standard deviation for each numerical column. Then, we use these values to standardize the column.

# The standardized data is then printed to the console.

# Here is an example of how standardization can be used to compare values from different columns:

# Suppose we have two columns, "sales" and "profit". The "sales" column has a mean of 100 and a standard deviation of 20. The "profit" column has a mean of 50 and a standard deviation of 10.

# If we want to compare the sales and profit for a particular row, we can simply look at the standardized values for those columns. If the standardized sales value is higher than the standardized profit value, then we know that the sales value is relatively high compared to the other sales values in the dataset.

# Standardization can also be used to identify outliers. Outliers are values that are significantly different from the rest of the data.

# To identify outliers, we can look at the standardized values for each column. If a value is more than 3 standard deviations away from the mean, then it is considered to be an outlier.
```

```python
# prompt: Standardize the numerical columns using the formula: z=x-mu/sigma

numerical_columns = sales_data.select_dtypes(include=["int64", "float64"]).columns
print(f"Numerical columns: {numerical_columns.tolist()}")

# Standardize the numerical columns
for column in numerical_columns:
  mean = sales_data[column].mean()
  std_dev = sales_data[column].std()
  sales_data[column] = (sales_data[column] - mean) / std_dev

# Print the standardized data
print(sales_data.head())
```

```python
# prompt: ​Show before and after comparisons of the data distributions.

import matplotlib.pyplot as plt

# Create histograms for each numerical column before and after standardization

for column in numerical_columns:
  # Before standardization
  plt.figure(figsize=(8, 4))
  plt.hist(sales_data[column], bins=10, edgecolor="black")
  plt.title(f"Distribution of {column} before standardization")
  plt.show()

  # After standardization
  plt.figure(figsize=(8, 4))
  plt.hist(sales_data[column], bins=10, edgecolor="black")
  plt.title(f"Distribution of {column} after standardization")
  plt.show()
```

```python
# prompt: Reflect on the importance of data preprocessing steps like standardization and one-hot encoding in data analysis and machine learning.

# Data preprocessing steps like standardization and one-hot encoding are essential for data analysis and machine learning because they help to:

# - Improve the accuracy and performance of machine learning models
# - Make it easier to interpret the results of data analysis
# - Identify and correct for potential biases in the data

# Standardization

# Standardization is a technique for scaling numerical columns to have a mean of 0 and a standard deviation of 1. This makes it easier to compare values from different columns and to identify outliers.

# Standardization is particularly important for machine learning models that are sensitive to the scale of the input data. For example, linear regression models and neural networks can be significantly affected by the scale of the input data.

# One-hot encoding

# One-hot encoding is a technique for converting categorical data into a numerical format that can be understood by machine learning models.

# One-hot encoding is particularly important for machine learning models that cannot handle categorical data directly. For example, linear regression models and neural networks cannot handle categorical data directly.

# By performing data preprocessing steps like standardization and one-hot encoding, we can improve the accuracy and performance of our machine learning models, make it easier to interpret the results of our data analysis, and identify and correct for potential biases in the data.

# Here are some specific examples of how standardization and one-hot encoding can be used in data analysis and machine learning:

# - Standardization can be used to compare the sales of different products, even if the products have different prices.
# - One-hot encoding can be used to identify the most important factors that contribute to customer churn.
# - Standardization and one-hot encoding can be used together to build a machine learning model that predicts the likelihood of a customer making a purchase.

# By taking the time to perform data preprocessing steps like standardization and one-hot encoding, we can improve the quality of our data analysis and machine learning results.
```
