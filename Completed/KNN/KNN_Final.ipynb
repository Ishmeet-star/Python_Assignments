```python
def check_missing_values(dataframe_name):
    for column in dataframe_name.columns:
      if dataframe_name[column].isnull().any():
        if dataframe_name[column].dtype == 'object':
          # Fill missing values in categorical columns with the most frequent value
          dataframe_name[column] = dataframe_name[column].fillna(dataframe_name[column].mode()[0])
        elif dataframe_name[column].dtype in ['int64', 'float64']:
          # Fill missing values in numerical columns with the mean
          dataframe_name[column] = dataframe_name[column].fillna(dataframe_name[column].mean())
    print(dataframe_name.isnull().sum())
    
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import warnings
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
warnings.filterwarnings("ignore")

training_accuracy_list = []
test_accuracy_list = []
training_precision_list = []
test_precision_list = []
training_recall_list = []
test_recall_list = []
training_f1_list = []
test_f1_list = []

for k in range(5, 18, 2):
    training_accuracy = []
    test_accuracy = []
    training_precision = []
    test_precision = []
    training_recall = []
    test_recall = []
    training_f1 = []
    test_f1 = []

    for i in range(1, 100, 1):
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=i)
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, Y_train)
        y_pred_train = knn.predict(X_train)
        y_pred_test = knn.predict(X_test)
        
        training_accuracy.append(accuracy_score(Y_train, y_pred_train))
        test_accuracy.append(accuracy_score(Y_test, y_pred_test))
        training_precision.append(precision_score(Y_train,y_pred_train, average='macro'))
        test_precision.append(precision_score(Y_test, y_pred_test,average='macro'))
        training_recall.append(recall_score(Y_train,y_pred_train, average='macro'))
        test_recall.append(recall_score(Y_test, y_pred_test,average='macro'))
        training_f1.append(f1_score(Y_train,y_pred_train, average='macro'))
        test_f1.append(f1_score(Y_test, y_pred_test,average='macro'))
        

    training_accuracy_list.append(np.mean(training_accuracy).round(2))
    test_accuracy_list.append(np.mean(test_accuracy).round(2))
    training_precision_list.append(np.mean(training_precision).round(2))
    test_precision_list.append(np.mean(test_precision).round(2))
    training_recall_list.append(np.mean(training_recall).round(2))
    test_recall_list.append(np.mean(test_recall).round(2))
    training_f1_list.append(np.mean(training_f1).round(2))
    test_f1_list.append(np.mean(test_f1).round(2))


print("Training Accuracies:", training_accuracy_list)
print("Test Accuracies:", test_accuracy_list)
print("Training Precision:", training_precision_list)
print("Test Precision:", test_precision_list)
print("Training Recall:", training_recall_list)
print("Test Recall:", test_recall_list)
print("Training F1:", training_f1_list)
print("Test F1:", test_f1_list)

```
